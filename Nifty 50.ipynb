**Seprated Banks Data For Further Analysis**

```python
banks= df['Symbol'].isin(['AXISBANK','HDFCBANK','ICICIBANK','INDUSINDBK','KOTAKBANK','SBIN','HDFC','BAJAJFINSV','BAJFINANCE','HDFCLIFE'])
bankdata = df[banks].pivot_table(values =['Open','High','Low','LTP','Chng','Volume (lacs)','52w H','52w L','Turnover (crs.)','365 d % chng','30 d % chng'], index= ['Symbol'])
bankdata.reset_index( inplace=True)
bankdata
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)
sns.barplot(data = bankdata, x = 'Symbol', y = '30 d % chng',palette='mako')
```

AxisBank and IndusInd Bank are the worst performers with more than 20% loss of the investors in a single month

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)
sns.barplot(data = bankdata, x = 'Symbol', y = '365 d % chng',palette='mako')
plt.xlabel("Symbol")  
plt.ylabel('% Change in a Year')  
plt.title("Yearly Growth")  
plt.show
```

BajajFinsv, BajajFinance,IciciBank,SBIN outperformed in Banking sector, giving more than 40% returns in a Year
